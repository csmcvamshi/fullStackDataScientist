```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

cols = ["CRIM","ZN","INDUS","CHAS","NOX","RM","AGE","DIS","RAD","TAX","PTRATIO","B","LSTAT","MEDV"]
df = pd.read_csv("../housing.csv", header=None, names=cols)

# Convert all columns to numeric (coerce errors to NaN)
df = df.apply(pd.to_numeric, errors="coerce")

# Drop rows with NaN if any
df = df.dropna()

print(df.head())
print(df.info())
print(df.describe())

arr = df.to_numpy()
print(np.mean(arr, axis=0))
print(np.std(arr, axis=0))
print(np.percentile(arr, [25, 75], axis=0))

df["RM_LSTAT_Ratio"] = df["RM"] / df["LSTAT"]
df["Tax_to_Rad"] = df["TAX"] / df["RAD"]

sns.set(style="whitegrid", palette="muted")

plt.figure(figsize=(10, 5))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()

plt.figure(figsize=(8, 5))
sns.pairplot(df[["RM","LSTAT","PTRATIO","MEDV"]])
plt.show()

plt.figure(figsize=(8, 5))
sns.scatterplot(x="RM", y="MEDV", data=df)
plt.figure(figsize=(8, 5))
sns.scatterplot(x="LSTAT", y="MEDV", data=df)
plt.figure(figsize=(8, 5))
sns.histplot(df["MEDV"], bins=30, kde=True)
plt.figure(figsize=(8, 5))
sns.boxplot(data=df[["CRIM","ZN","INDUS","NOX","AGE","TAX"]])
plt.show()
```

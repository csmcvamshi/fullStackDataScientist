# seaborn - statistical data visualisation

### what is seaborn   

- seaborn is a py lib built on top of matplotlib that makes the statistical data more easy and beautiful

### why we need it?

- for creating plots like bionic plot , box plot and plain plots
- automatically handles data from pandas dataframe

### where we use it?

- data analysis and reporting
- exploring relations between variables
- ML (visualize datasets and results)
    

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("tested.csv")
print(df.describe())
print(df.info())
print(df.shape)
```

```python
## 4. Histograms & Countplots
# Histogram
sns.histplot(df['Age'].dropna(), bins=10, kde=True)
plt.title("Histogram of Age")
plt.show()
 
# Countplot
sns.countplot(x='Pclass', data=df)
plt.title("Count of Passengers by Class")
plt.show()
```

```python
### 5. Scatter Plots
sns.scatterplot(x='Age', y='Fare', data=df, hue='Sex')
plt.title("Age vs Fare")
plt.show()
```

```python
### 6. Boxplot & Violin Plot
# Boxplot
sns.boxplot(x='Pclass', y='Age', data=df)
plt.title("Boxplot of Age by Class")
plt.show()
 
# Violin plot
sns.violinplot(x='Pclass', y='Age', hue='Sex', data=df, split=True)
plt.title("Violin Plot of Age by Class and Sex")
plt.show()
```

```python
### 7. Pairplot
iris = sns.load_dataset('iris')
sns.pairplot(iris, hue='species')
plt.show()
```

```python
sns.pairplot(df)
plt.show()
```

```python

sns.heatmap(df.corr(numeric_only=True), annot=True)
plt.title("Correlation Heatmap")
plt.show()
```

```python
### 9. Customization & Themes
sns.set_theme(style="whitegrid")  # options: darkgrid, white, ticks, etc.
sns.histplot(df['Age'].dropna(), bins=10, color='orange')
plt.title("Customized Histogram with Seaborn")
plt.show()
```

```python
sns.barplot(x='Pclass', y='Fare', data=df, hue='Sex', ci='sd')
plt.title("Average Fare by Class and Sex")
plt.show()
```

```python
titanic = sns.load_dataset("titanic")
sns.catplot(data=titanic, x="pclass", hue="sex", col="survived",
            kind="count", height=4, aspect=0.8)
plt.show()
```

```python
sns.histplot(data=titanic, x="age", kde=True, bins=30)
plt.title("Age Distribution of Passengers")
plt.show()
```

```python
sns.boxplot(data=titanic, x="pclass", y="fare")
plt.title("Fare Distribution by Class")
plt.show()
```

```python
sns.scatterplot(data=titanic, x="age", y="fare", hue="pclass")
plt.title("Age vs Fare by Class")
plt.show()
```

```python
corr = titanic.select_dtypes(include=["number"]).corr()
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```
